```python
# CRIAR CÓDIGO QUE DADO A PÁGINA RETORNE E ORDENE COM BASE NO PREÇO OS NOTEBOOKS DA MARCA LENOVO

from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import csv


driver = webdriver.Chrome()
driver.get("https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops")

# Atribuir o conjunto de elementos da página as variaveis
Notebooks = driver.find_elements(By.CLASS_NAME,"title") #TITULO NOTEBOOKS
Precos = driver.find_elements(By.CLASS_NAME,"price") # PREÇO NOTEBOOKS
Descricao = driver.find_elements(By.CLASS_NAME,"description") # DESCRICAO

#LISTAS QUE IRÃO RECEBER OS RESPECTIVOS DADOS OBTIDOS
listaNot = []
ListaPrecos = []
ListaDescricao = []

#FOR QUE IRÁ PERCORRER A LISTA E ADICIONAR A INFORMAÇÃO ÀS VARIAVEIS

for i in range(len(Notebooks)):
    
    if "Lenovo" in Notebooks[i].text : # CONDIÇÃO DE APENAS ADICIONAR SE FOR DA MARCA LENOVO
        listaNot.append(Notebooks[i].text)
        ListaPrecos.append(float((Precos[i].text).replace("$",""))) # O ".REPLACE" é para remover o $ da STRING e o "FLOAT" para converter a str em float
        ListaDescricao.append(Descricao[i].text)


tabela = {} # Dicionario que receberá todos os dados de acordo as chaves declaradas abaixo

tabela["Modelo"] = listaNot
tabela["Preço"] = ListaPrecos
tabela["Descrição"] = ListaDescricao

# Método do Pandas para ordenar pelo preço e exibir de forma organizada 
tabela_notebook = pd.DataFrame.from_dict(tabela)
tabela_ordenada = tabela_notebook.sort_values(by=['Preço'],ignore_index=True)

display(tabela_ordenada)

#Arquivo salvo no formato CSV para posterior utilização
tabela_ordenada.to_csv("lenovoByPrice.csv", index = False)


```
